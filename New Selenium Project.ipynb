```python
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium import webdriver
import pandas as pd
import time


webdriver_service = Service('path/to/chromedriver')  # Replace with the path to your chromedriver executable
webdriver_options = Options()
driver = webdriver.Chrome(service=webdriver_service, options=webdriver_options)
driver.get("https://www.saucedemo.com/inventory.html")

time.sleep(2)

User_Name = driver.find_element(By.ID, "user-name")
User_Name.send_keys("standard_user")

time.sleep(2)
Password = driver.find_element(By.ID, "password")
Password.send_keys("secret_sauce")

time.sleep(2)

Login_button = driver.find_element(By.ID, "login-button")
Login_button.click()

name = driver.find_elements(By.XPATH, "//div[@class='inventory_item_name']")
Name = []
for i in name:
    Name.append(i.text)
    
detail = driver.find_elements(By.XPATH, "//div[@class='inventory_item_desc']")
Details = []
for i in detail:
    Details.append(i.text)
    
price = driver.find_elements(By.XPATH, "//div[@class='inventory_item_price']")
Price = []
for i in price:
    Price.append(i.text)

df = pd.DataFrame(zip(Name,Details,Price),columns=['Name','Detailse','Price'])
df
```
